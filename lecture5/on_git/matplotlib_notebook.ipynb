# Matplotlib for visualisation

### What is matplotlib?

 * Matplotlib is a Python package for generating 2D and 3D plots
 * Matlab like plotting library. Follows naming convention/formats as in MATLAB
 * Very popular. It is extremely flexible - all aspects of plots can be controlled programmatically
 * Can be used to draw graphs, barcharts, scatter plots, contour and surface plots etc
 * Can produce very high quality output in numerous formats (png, pdf, svg etc)
 
### Official website: http://matplotlib.org/

 * Guidelines for installing the library: https://matplotlib.org/users/installing.html

```python
# First thing to do: import the necessary libraries
# STEP 1

import matplotlib.pyplot as plt     # plt is an alias that can now be used to access the functionalities of pyplot
import numpy as np

#%matplotlib inline    
# enabling this, we don't have to write plt.show() everytime, graphs/plots are displayed inline automatically
```

# Examples

## Plotting Numpy function

```python
# The most important function in matplotlib is plot, which allows you to plot 2D data. Here is a simple example:
# Compute the x and y coordinates for points on a sine curve

#STEP2 : prepare the data for plotting
    
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# If you want to see how your x and y data looks like, just enable the print statements below
#print(x)
#print(y)
```

```python
# Step 3: plot the points calling plot function and passing the necessary parameters

plt.plot(x, y, 'b')    #x,y represents 2D data
                       #third parameter 'b' is used to chose the color


# Step 4: show the plots by calling show() function to make graphics appear.
plt.show()              
         # However, if i had enabled %matplotlib inline  in the beginning then I don't need to
         # call plt.show() to see the plots. Try yourself.
```

## Controlling the line attributes

```python
# Playing around with: color, linewidth, linestyle parameters of the plot() function

# use blue color, plot with two pixel linewidth
y_data=np.sin(x)
plt.plot(x,y_data,color='blue',linewidth=2.0, linestyle='-')
     
# use red color, plot with two pixel linewidth, dot separated lines
y_data=np.cos(x)     
plt.plot(x,y_data,color='red',linewidth=2.0, linestyle='-.')
     
# plot the grid
plt.grid(True)


# note: when we call plot function multiple times, the plots are plotted on the same figure
```

### Adding labels, constraining the plotting range

```python
# We can use xlim(x1, x2), ylim(y1,y2) to constrain the plotting range

x=np.arange(-5,5,0.1)
y=x**2

# print the data x and y  . Uncomment if you wish to see how your x and y looks like
#print(x)
#print(y)

plt.xlim(-5,5)   # Set the x axis range
plt.ylim(0,100)  # Set the y axis range

plt.xlabel("x")     # Set the x axis label
plt.ylabel("y=x*x") # Set the y axis label
plt.title("Plot y=x*x") # Set the title for the plotting

plt.plot(x,y)
plt.show()
```

## Set the ticks of axes

```python
# We can specify the ticks of axes xticks (yticks)

#set the data
x=np.linspace(-np.pi,np.pi,100)
y_data = np.cos(x)

plt.xticks(np.linspace(-np.pi,np.pi,5))

plt.plot(x,y_data,color='red',linewidth=2.0,linestyle='-')
         
plt.show()
```

## Set the legends

```python
# To add legends in your plots, we make use of the label parameter
# and assign appropriate labels to them

x=np.linspace(-np.pi,np.pi,100)
y_sin,y_cos = np.sin(x), np.cos(x)     # python allows to write two statements in a single line

plt.plot(x,y_sin,color='blue',linewidth=2.0, linestyle='-',label='sin')
plt.plot(x,y_cos,color='red',linewidth=2.0, linestyle='-',label='cos')

plt.title('Sine and Cosine')
plt.legend(loc='upper left')
plt.show()
```

# Subplots

 * We can plot different things in the same figure using the subplot function
 * subplot() function
 * 3 integer parameters. The first two indicate the total number of rows and columns. 
 * The last integer indicates the index of this subplot
 

```python
# Compute the x and y coordinates for points on sine and cosine curves

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()

# Note: here the two subplots arranged vertically
```

```python
# Two subplots arragned horizontally

x = np.linspace(0, 5, 10)
y = x ** 2

plt.subplot(1,2,1)
plt.plot(x, y, 'r--')

plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')

plt.show()
```

# Pie chart

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:

labels = 'Maths', 'Biology', 'Chemistry', 'Physics'

data = [15, 30, 45, 10] # percentile distribution
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(data, explode=explode, labels=labels,autopct='%1.1f%%',startangle=90)
plt.show()

# autopct='%1.1f%% allows the proportion of numbers appear in the respectives slices
```

# Scatter plots

```python
# scatter function plot points

n=100
x=np.random.normal(0,1,n)
y=np.random.normal(0,1,n)

plt.scatter(x,y)
plt.show()
```

# Bar plots

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5'] # lets assume we have 5 different groups 
                                        # each group data represented by mean value for both genders
 
men_means = [20, 34, 30, 35, 27]    
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels)) # the label locations

width = 0.35 # the width of the bars
rects1 = plt.bar(x - width/2, men_means, width,label='Men')
rects2 = plt.bar(x + width/2, women_means, width,label='Women')

plt.legend()

plt.show()
```

# Figure size and resolution

You can indicate the resolution, size, and aspect ratio of a plot

 * To create an 800*600, 100dpi resolution image, you can use the following command
   
    fig, axes = plt.subplots(figsize=(8,6),dpi=100) 

```python
x = np.linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(figsize=(8,6),dpi=100)

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')

plt.show()
```

# Multiple axes in a figure

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes1.plot(x, y, 'r-')
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
axes2.plot(x,-y, 'g-')
```

# Using the subplot layout manager

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
for ax in axes.reshape(-1):  # axes is returned as a 2x3 numpy array; reshape into a vector
    ax.plot(x,y,'r')
    ax.set_ylabel('y axis')
    ax.set_xlabel('x axis')
    
fig.tight_layout()            # adjusts spacing to avoid plots overlapping labels
```

# Adding text in your plots

```python
x = np.linspace(-5,5,10)
y = x*2+1
plt.plot(x,y)

plt.axis('tight')
plt.text(0, -2, 'The line is blue', fontsize=16)
plt.text(-2, 5, '$y=2x+1$', fontsize=16)

plt.show()
```

# 3D figures

 * For 3D plot, we need matplotlib library 
 * and mpl_toolkits.mplot3d library
 

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(8,4))

# prepare 3D data
axis=np.linspace(-5,5,100)
(X,Y) = np.meshgrid(axis,axis)    # outputs matrices X and Y storing grid of pixel coordinates
Z = np.cos(X) + np.sin(Y)        # perform some operation on the coordinate values

ax1 = fig.add_subplot(1,2,1, projection='3d') # generate an Axes3D
im = ax1.plot_surface(X,Y,Z, cmap=plt.cm.gray)


ax2 = fig.add_subplot(1,2,2, projection='3d') # generate an Axes3D
im = ax2.plot_wireframe(X,Y,Z, rstride=10, cstride=10)
```
